```python
# Import required libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load Titanic dataset from seaborn library
df = sns.load_dataset("titanic")
```

```python
# Display first 5 rows of dataset
df.head()
```

```python
# Display column names
df.columns
```

```python
# Display dataset information (data types, null values, etc.)
df.info()
```

```python
# Check number of missing values in each column
df.isnull().sum()
```

```python
# Drop unnecessary columns from dataset
df.drop(["deck","embark_town","alive","class","who","adult_male"],axis=1,inplace=True)
```

```python
# Check dataset information again after dropping columns
df.info()
```

```python
# Fill missing values in 'age' column using mean value
df['age'].fillna(df['age'].mean(),inplace=True)
```

```python
# Remove rows where 'embarked' column has missing values
df.dropna(subset=['embarked'],inplace=True)
```

```python
# Check dataset info after handling missing values
df.info()
```

```python
# Import LabelEncoder to convert categorical data into numeric values
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
# Convert 'sex' and 'embarked column into numeric form (male=1, female=0)
df['sex']=le.fit_transform(df['sex'])
df['embarked']=le.fit_transform(df['embarked'])
```

```python
# Convert all columns to integer type
df=df.astype(int)
df.head()
```

```python
# Separate input features (X) and target variable (y)
X=df.drop('survived',axis=1)
y=df['survived']
```

```python
# Split dataset into training and testing sets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size=0.20, random_state=42)
```

## **Logistic Regression**

```python
from sklearn.linear_model import LogisticRegression
# Create Logistic Regression model
model=LogisticRegression()
# Train model
model.fit(X_train,y_train)
```

```python
# Predict on test data
y_pred=model.predict(X_test)
y_pred
```

```python
# Actual test values
y_test
```

```python
# Import evaluation metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
# Calculate accuracy
accuracy_score(y_test,y_pred)
```

```python
# Generate confusion matrix
confusion_matrix(y_test,y_pred)
```

```python
# Generate classification report (precision, recall, f1-score)
print(classification_report(y_test,y_pred))
```

```python
#Feature Scaling
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
X_test_scaled=scaler.transform(X_test)
```

## **K-Nearest Neighbour**

```python
from sklearn.neighbors import KNeighborsClassifier
# Create KNN model with 5 neighbors
knn = KNeighborsClassifier(n_neighbors=5)
# Train the model
knn.fit(X_train_scaled, y_train)
```

```python
# Predict test data
y_knn_pred=knn.predict(X_test_scaled)
```

```python
# Evaluate model accuracy
accuracy_score(y_test,y_knn_pred)
```

```python
# Confusion matrix
confusion_matrix(y_test,y_knn_pred)
```

```python
# Classification report
print(classification_report(y_test,y_knn_pred))
```

## **Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
# Create Naive Bayes model
nb=GaussianNB()
# Train the model
nb.fit(X_train,y_train)
```

```python
# Predict test data
y_pred_NB = nb.predict(X_test)
y_pred_NB
```

```python
accuracy_score(y_test,y_pred_NB)
```

```python
print(classification_report(y_test,y_pred_NB))
```

# **Decision trees**

```python
from sklearn.tree import DecisionTreeClassifier
# Create Decision Tree model
dt=DecisionTreeClassifier(random_state=42)
# Train the model
dt.fit(X_train_scaled,y_train)
```

```python
# Predict test data
y_pred_dt=dt.predict(X_test_scaled)
y_pred_dt
```

```python
# Classification report
print(classification_report(y_test,y_pred_dt))
```

```python
accuracy_score(y_test,y_pred_dt)
```

## **SVM**

```python
from sklearn.svm import SVC
```

```python
# Create SVM model with RBF kernel
model_SVM=SVC(kernel='rbf')
# Train the model
model_SVM.fit(X_train_scaled,y_train)
```

```python

# Predict test data
y_pred_SVM=model_SVM.predict(X_test_scaled)
y_pred_SVM
```

```python
accuracy_score(y_test,y_pred_SVM)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
# Perform 5-fold cross validation using SVM model
scores = cross_val_score(model_SVM, X_scaled, y, cv=5,scoring='accuracy')
```
